```python
Ks = np.logspace(-3, 3, 10)
```

```python
mean_exp = []
std_exp = []
mean_bound = []
std_bound = []
mean_v2 = []
for K in tqdm(Ks):
    exp1 = ConstantExperiment(exp.N, exp.P[1:], K, exp.W, exp.B, activation = 'relu')
    mean_exp_, std_exp_, mean_bound_, std_bound_, _, _, mean_v2_ = exp1.run(repetitions = 1000, inputs = 50, do_tqdm = False, do_print = False, do_plot = False)
    mean_exp += [mean_exp_]
    std_exp += [std_exp_]
    mean_bound += [mean_bound_]
    std_bound += [std_bound_]
    mean_v2 += [mean_v2_]
```

```python
plt.figure()
plt.title('Error comparison for different K and fixed weights')
plt.xlabel('K, Lipschitz coefficient')
plt.ylabel('Output error')
plt.plot(Ks, mean_exp, label = 'Experimental error (mean)')
plt.plot(Ks, mean_bound, label = 'Erf error (mean)')
plt.plot(Ks, np.mean(mean_v2, axis=1), label = 'Erfv2 error (mean)')
plt.plot(Ks, [1] * len(mean_exp), label = 'y=1')
plt.legend()
plt.xscale('log')
plt.yscale('log')
plt.show()
```
