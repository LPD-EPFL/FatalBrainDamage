```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
```

```python
# taken from https://www.math.ubc.ca/~pwalls/math-python/integration/riemann-sums/
f = lambda x : 1/(1+(np.sin(5 * x)-0.5)**2)
a = 0; b = 1; N = 10
n = 10 # Use n*N+1 points to plot the function smoothly

plt.figure(figsize=(10,2))
def plot_N(N = 10):
    
    """ Plot for a fixed N neurons """
    
    x = np.linspace(a,b,N+1)
    y = f(x)

    X = np.linspace(a,b,n*N+1)
    Y = f(X)
    
    plt.plot(X,Y,'b')
    x_left = x[:-1] # Left endpoints
    y_left = y[:-1]
    plt.plot(x_left,y_left,'b.',markersize=10, label = 'Continuous z(t_l), $t_l\in[0,1]$')
    plt.bar(x_left,y_left,width=(b-a)/N,alpha=0.2,align='edge',edgecolor='b', label = 'Discrete $z^{i_l}$, $n_l = %d$' % N)
    plt.ylabel('$z_l$')
    plt.legend()

plt.subplot(1,3,1)
plot_N(10)

plt.subplot(1,3,2)
plot_N(20)

plt.subplot(1,3,3)
plot_N(30)

plt.savefig('continuos_net.pdf', bbox_inches = 'tight')
    
plt.show()
```
