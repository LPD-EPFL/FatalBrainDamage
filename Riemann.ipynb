```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
import matplotlib
```

```python
font = {'family' : 'normal',
        'weight' : 'normal',
        'size'   : 20}
matplotlib.rc('font', **font)
```

```python
# taken from https://www.math.ubc.ca/~pwalls/math-python/integration/riemann-sums/
f = lambda x : 1/(1+(np.sin(5 * x)-0.5)**2)
a = 0; b = 1; N = 10
n = 10 # Use n*N+1 points to plot the function smoothly

plt.figure(figsize=(7,1))
def plot_N(N = 10):
    
    """ Plot for a fixed N neurons """
    
    x = np.linspace(a,b,N+1)
    y = f(x)

    X = np.linspace(a,b,n*N+1)
    Y = f(X)
    
    plt.plot(X,Y,'b')
    x_left = x[:-1] # Left endpoints
    y_left = y[:-1]
    line = plt.plot(x,y,'b.',markersize=10, label = 'Cont. $z(t)$, $t\in[0,1]$')
    bar = plt.bar(x_left,y_left,width=(b-a)/N,alpha=0.2,align='edge',edgecolor='b', label = 'Discrete $z^{i}$, $i\in[n]$')
    #plt.ylabel('$z_l$')
    plt.ylim(0.2,1.2)
    #plt.legend()
    plt.xlabel('($\mathit{n = %d}$)' % N, fontsize=15)
    cur_axes = plt.gca()
    cur_axes.axes.get_xaxis().set_label_coords(0.5,-0.1)
    if N > 10:
        #cur_axes.axes.get_xaxis().set_visible(False)
        cur_axes.axes.get_yaxis().set_visible(False)
    return line, bar

plt.subplot(1,3,1)
plot_N(10)

plt.subplot(1,3,2)
plot_N(20)

plt.subplot(1,3,3)
line, bar = plot_N(30)

plt.legend(loc="upper left", bbox_to_anchor=(1,1.2))
plt.savefig('figures/continuous_net.pdf', bbox_inches = 'tight')
    
plt.show()
```
