```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
from experiment_random import *
from tqdm import tqdm
```

```python
np.random.seed(42)
P = 0.123
K = 0.321
RandomExperiment(N = [10, 5, 5, 1], P = P, KLips = K, do_print = True).run(inputs = 20, repetitions = 10000, do_print = True, do_plot = True, do_tqdm = True)
```

```python
K = 0.1
Ps = np.linspace(0, 0.9, 20)
results = [RandomExperiment(P = [P] * 1, N = [10, 5, 1], KLips = K).run(inputs = 20, do_tqdm = False, do_plot = False, do_print = False) for P in tqdm(Ps)]
```

```python
mean_exp, std_exp, mean_bound, std_bound, std_true = [np.array(x) for x in zip(*results)]
plt.figure()
plt.title('Errors w.r.t. P')
plt.xlabel('P')
plt.ylabel('Network absolute error')
plt.plot(Ps, mean_exp, label = 'Experimental mean')
plt.fill_between(Ps, mean_exp - std_exp, mean_exp + std_exp, alpha = 0.3)
plt.plot(Ps, mean_bound, label = 'Theoretical mean')
plt.fill_between(Ps, mean_bound - std_bound, mean_bound + std_bound, alpha = 0.3)
plt.legend()
plt.show()
```
